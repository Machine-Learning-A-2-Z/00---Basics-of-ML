````python
# 1. How to import
import numpy as np
````

````python
# 2. creating numpy array
arr = np.array([1, 2, 3, 4, 5])
print(arr)
````

````python
# 3. creating array with zeros
arr = np.zeros((2, 3))
print(arr)
````

````python
# 4. creating array with ones
arr = np.ones((2, 3))
print(arr)
````

````python
# 5. creating array with range of numbers
arr = np.arange(10)
print(arr)
````

````python
# 6. Reshape an array
arr = np.array([1, 2, 3, 4, 5, 6])
new_arr = arr.reshape(2, 3)
print(new_arr)
````

````python
# 7. Basic arithmatic operation
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
print(a - b)
print(a * b)
print(a / b)
````

````python
# 8. Matrix multiplication
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
print(np.dot(matrix1, matrix2))
````

````python
# 9. Finding Mean and Standard Deviation
arr = np.array([1, 2, 3, 4, 5])
print(np.mean(arr))
print(np.std(arr))
````

````python
# 10. Transposing a matrix
matrix = np.array([[1, 2], [3, 4]])
print(matrix.T)
````

````python
# matplotlib.pyplot to draw graph
import matplotlib.pyplot as plt
````

````python
# 2D
arr = np.array([2, 1, 5, 7, 4, 6, 8, 14, 10, 9, 18, 20, 22])

plt.plot(arr)
````

````python
# 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
````

````python
# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.arange(10)
y = np.arange(10)
z = x + y
# Scatter plot of x, y, z (since z is a scalar, you can't plot it directly as a curve)
ax.scatter(x, y, z)  # Example, plotting x * y as z values

# Add title and labels
ax.set_title('3D Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis (x+Y)')

# Show the plot
plt.show()
````
