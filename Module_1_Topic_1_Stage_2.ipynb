````python
import seaborn as sns
import matplotlib.pyplot as plt
'''
Me: Seaborn is a Python data visualization library based on matplotlib.
It provides a high-level interface for drawing attractive and informative statistical graphic
You: Ok
'''
````

````python


# Generate pairplot for multivariate analysis
sns.pairplot(df)
plt.show()

'''
Me: pair plot help to visualize relationship between any two variables
You: Perfect
'''
````

````python
# Generate correlation matrix
correlation_matrix = df.corr()
print("Correlation Matrix:\n", correlation_matrix)
'''
Me: Correlation describes the strength of an association between two variables
You: Can we plot it ?

Me: Yes

'''
````
