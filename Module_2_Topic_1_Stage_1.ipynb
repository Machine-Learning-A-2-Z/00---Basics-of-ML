````python
arr = np.array([2], dtype = np.int32)
brr = np.array([2.0], dtype = np.float64)

crr = arr + brr
print(f"crr = {crr}")
print(f"dtype(crr) = {crr.dtype}")
````

# **Copies vs Views**



```
Me: Ever wonder why changing an array sometimes affects the original and sometimes not?

You: explain problem
```



````python
arr = np.array([1, 2, 3, 4, 5])
copy = arr.copy()
arr[0] = 10000  # modify at index 0

print(f"arr: {arr}")
print(f"copy: {copy}")
````

````python
arr = np.array([1, 2, 3, 4, 5])
view = arr
arr[0] = 10000  # modify at index 0

print(f"arr: {arr}")
print(f"view: {view}")
````
