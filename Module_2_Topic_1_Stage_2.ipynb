### 1. **Basic Arithmetic Operations**
   These are element-wise operations that work on arrays of the same size.
   - **Addition**: `np.add(x, y)` or `x + y`
   - **Subtraction**: `np.subtract(x, y)` or `x - y`
   - **Multiplication**: `np.multiply(x, y)` or `x * y`
   - **Division**: `np.divide(x, y)` or `x / y`
   - **Modulus**: `np.mod(x, y)` or `x % y`
   - **Power**: `np.power(x, y)` or `x ** y`

### 2. **Matrix Operations**
   - **Dot Product**: `np.dot(a, b)` or `a @ b`
   - **Matrix Multiplication**: `np.matmul(a, b)`
   - **Transpose**: `np.transpose(a)` or `a.T`
   - **Matrix Determinant**: `np.linalg.det(a)`
   - **Inverse Matrix**: `np.linalg.inv(a)`
   - **Eigenvalues and Eigenvectors**: `np.linalg.eig(a)`

### 3. **Statistical Operations**
   - **Mean**: `np.mean(a)`
   - **Median**: `np.median(a)`
   - **Standard Deviation**: `np.std(a)`
   - **Variance**: `np.var(a)`
   - **Min/Max**: `np.min(a)`, `np.max(a)`
   - **Sum of Elements**: `np.sum(a)`
   - **Cumulative Sum**: `np.cumsum(a)`
   - **Product of Elements**: `np.prod(a)`
   - **Cumulative Product**: `np.cumprod(a)`

### 4. **Linear Algebra Operations**
   - **Solving Linear Equations**: `np.linalg.solve(a, b)`
   - **QR Decomposition**: `np.linalg.qr(a)`
   - **Singular Value Decomposition (SVD)**: `np.linalg.svd(a)`
   - **Cholesky Decomposition**: `np.linalg.cholesky(a)`

### 5. **Trigonometric Functions**
   - **Sine**: `np.sin(x)`
   - **Cosine**: `np.cos(x)`
   - **Tangent**: `np.tan(x)`
   - **Inverse Sine**: `np.arcsin(x)`
   - **Inverse Cosine**: `np.arccos(x)`
   - **Inverse Tangent**: `np.arctan(x)`
   - **Hyperbolic Sine**: `np.sinh(x)`
   - **Hyperbolic Cosine**: `np.cosh(x)`
   - **Hyperbolic Tangent**: `np.tanh(x)`

### 6. **Exponential and Logarithmic Functions**
   - **Exponential**: `np.exp(x)`
   - **Natural Logarithm (base e)**: `np.log(x)`
   - **Logarithm (base 10)**: `np.log10(x)`
   - **Logarithm (base 2)**: `np.log2(x)`

### 7. **Rounding Operations**
   - **Round**: `np.round(x)`
   - **Floor**: `np.floor(x)`
   - **Ceil**: `np.ceil(x)`

### 8. **Set Operations**
   - **Union**: `np.union1d(a, b)`
   - **Intersection**: `np.intersect1d(a, b)`
   - **Difference**: `np.setdiff1d(a, b)`
   - **Unique Elements**: `np.unique(a)`

### 9. **Broadcasting and Scalar Operations**
   - NumPy allows broadcasting, enabling operations between arrays of different shapes if compatible. For instance:
     ```python
     a = np.array([1, 2, 3])
     b = 2
     result = a * b  # Multiplies each element of 'a' by 2
     ```

### 10. **Sorting and Searching**
   - **Sorting Arrays**: `np.sort(a)`
   - **Find Indices of Sorted Elements**: `np.argsort(a)`
   - **Searching for Elements**: `np.where(condition)`
   - **Find Maximum/Minimum Index**: `np.argmax(a)`, `np.argmin(a)`

### 11. **Random Number Generation**
   - **Random Array**: `np.random.rand(n)`
   - **Random Integer**: `np.random.randint(low, high, size)`
   - **Normal Distribution**: `np.random.normal(mean, std, size)`
   - **Shuffle Array**: `np.random.shuffle(a)`

### 12. **Logical Operations**
   - **Element-wise Logical AND**: `np.logical_and(a, b)`
   - **Element-wise Logical OR**: `np.logical_or(a, b)`
   - **Element-wise Logical NOT**: `np.logical_not(a)`
   - **Comparisons**: `np.greater(a, b)`, `np.less(a, b)`, `np.equal(a, b)`, etc.

### 13. **Fourier Transform**
   - **1D Fourier Transform**: `np.fft.fft(a)`
   - **Inverse Fourier Transform**: `np.fft.ifft(a)`

### 14. **Polynomials**
   - **Polynomial Fit**: `np.polyfit(x, y, degree)`
   - **Evaluate Polynomial**: `np.polyval(p, x)`

````python
transactions = [50, -20, 30, -10]
cumulative_balance = np.cumsum(transactions)

````

**Cummulative sum in finance and accounting**

````python
import matplotlib.pyplot as plt
````

````python


# Plotting
plt.figure(figsize=(8, 5))

# Plot line for cumulative balance
plt.plot(cumulative_balance, marker='o', color='blue', label="Cumulative Balance", zorder=2)

# Plot histogram-style vertical lines for each transaction index
plt.bar(range(len(cumulative_balance)), cumulative_balance, color='skyblue', alpha=0.6, label="Transaction Points", zorder=1)

# Adding labels and grid
for i, balance in enumerate(cumulative_balance):
    plt.text(i, balance + 1, f"{balance}", ha='center', color='black')

plt.title("Cumulative Balance with Transaction Index")
plt.xlabel("Transaction Index")
plt.ylabel("Cumulative Balance")
plt.grid(True)
plt.legend()

# Show plot
plt.show()
````

**Cummulative Prod in Financial Management**

````python
import numpy as np
import matplotlib.pyplot as plt

# Example transaction data
transactions = [1.05, 0.9, 1.1, 0.95]  # These could represent growth factors or returns

# Calculate cumulative product
cumulative_product = np.cumprod(transactions)

# Plotting
plt.figure(figsize=(10, 6))

# Plot line for cumulative product with larger markers
plt.plot(cumulative_product, marker='o', color='green', label="Cumulative Product", linewidth=2, markersize=8, zorder=2)

# Plot histogram-style vertical lines for each transaction index with distinct colors
plt.bar(range(len(cumulative_product)), cumulative_product, color='lightgreen', alpha=0.7, label="Transaction Points", zorder=1)

# Adding labels with spacing
for i, balance in enumerate(cumulative_product):
    plt.text(i, balance + 0.02, f"{balance:.2f}", ha='center', color='black', fontsize=10)

# Set y-axis limits for better label visibility
plt.ylim(min(cumulative_product) - 0.1, max(cumulative_product) + 0.2)

# Adding title and labels
plt.title("Cumulative Product with Transaction Index", fontsize=14)
plt.xlabel("Transaction Index", fontsize=12)
plt.ylabel("Cumulative Product", fontsize=12)

# Adding grid, legend, and showing plot
plt.grid(True)
plt.legend()
plt.show()
````
